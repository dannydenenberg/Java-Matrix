```python
# find k-anonymity of a dataset using groupby
def find_k_anonymity(df, quasi_identifiers):
    k_anonymity = df.groupby(quasi_identifiers).size().min()
    return k_anonymity

QUAZI_IDS = list(df.columns)
QUAZI_IDS.remove('course_id')
QUAZI_IDS.remove('user_id')

k_anonymity = find_k_anonymity(df, QUAZI_IDS)
print(f"The original dataset has a k-anonymity of {k_anonymity}.") # returns 1
```
